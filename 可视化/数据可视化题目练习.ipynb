## 绘制折线图

【例2-1】绘制中国银行2015-2019年营业收入折线图

要求： 1.设置x轴、y轴有标题；
     2.设置图表标题为“中国银行2015-2019年营业收入”；
     3.折线图要显示刻度
    
数据源：revenue.xlsx

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import pandas as pd

mpl.rcParams['font.sans-serif'] = ['FangSong']
mpl.rcParams['axes.unicode_minus'] = False

# 读取数据
df = pd.read_excel('D:\datasets\mldata\Revenue.xlsx', header=None)
data = df.values

# 处理数据
x = np.linspace(2015, 2019, 5)
y = data[1][1:]

plt.figure(figsize = (15, 7), dpi = 80) # 设置图像大小
plt.plot(x, y,
         marker='o')
# plt.xticks([2015, 2016, 2017, 2018, 2019])
plt.xticks(range(2015, 2020))
plt.xlabel('年份')  # x轴标签
plt.ylabel('营业收入(单位：亿)')  # y轴标签
plt.title('中国银行2015-2019年营业收入折线图')  # 标题
plt.show()
```

## 绘制柱形图

<font color=black size=3 face="微软雅黑">
    
【例2-4】绘制中国银行和建设银行2015-2019年营业收入柱状图
    
要求： 
  
   1.设置x轴、y轴有标题；
    
   2.设置图表标题为“中国银行和建设银行2015-2019年营业收入柱形图”；
    
   3.显示刻度
    
   4.设置图例显示
    
数据源：revenue.xlsx数据源：
    

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

mpl.rcParams['font.sans-serif'] = ['FangSong']
mpl.rcParams['axes.unicode_minus'] = False

data = pd.read_excel('E:\项目文件\Python\可视化文档\Revenue.xlsx').values

x1 = [i for i in range(len(data[0]) - 1)]
y1 = [data[0][i] for i in range(1, len(data[0]))]
x2 = [i + 0.2 for i in range(len(data[1]) - 1)]
y2 = [data[1][i] for i in range(1, len(data[1]))]

plt.figure(figsize = (20, 10), dpi = 80)

plt.bar(x1, y1, label = '中国银行', width = 0.2)
plt.bar(x2, y2, label = '建设银行', width = 0.2)

x = [i for i in range(2015, 2020)]
xticks = [i + 0.1 for i in x1]
plt.xticks(xticks, x)

plt.title('中国银行和建设银行2015-2019年营业收入柱形图')
plt.xlabel('年份')
plt.ylabel('年营业收入(单位:亿)')

plt.legend(loc = 'best')

for x, y in zip(x1, y1):
    plt.text(x, y + 50, '%s' %y, ha = 'center')
for x, y in zip(x2, y2):
    plt.text(x, y + 50, '%s' %y, ha = 'center')

plt.show()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Bar
import pandas as pd
import numpy as np

# 读取数据
df = pd.read_excel('E:\项目文件\Python\可视化文档\Revenue.xlsx', header=None)
data = df.values

# 处理数据
x = data[0][1:].tolist()
y1 = data[1][1:].tolist()  # 中国银行数据 列表
y2 = data[2][1:].tolist()  # 建设银行数据

# 绘制柱形图
bar = Bar()
bar.add_xaxis(x)
bar.add_yaxis('中国银行', y_axis=y1,category_gap = '50%', gap = '0') #category_gap宽度（百分比越大，宽度越小）   gap间隔
bar.add_yaxis('建设银行', y_axis=y2,category_gap = '50%', gap = '0')
bar.set_global_opts(xaxis_opts=opts.AxisOpts(name='年份'),
                    yaxis_opts=opts.AxisOpts(name='营业收入(单位：亿)'),  # 设置y轴刻度起点
                    title_opts=opts.TitleOpts(title='中国银行和建设银行2015-2019年营业收入柱形图', pos_left='center'),
                    legend_opts=opts.LegendOpts(type_='plain',
                                                is_show=True,
                                                pos_top='10%'
                                                ),
                    )

# bar.render(r"E:\项目文件\Python\render.html")
bar.render_notebook()
```

## 绘制直方图

<font color=black size=3 face="微软雅黑">
    
【例2-8】使用直方图分析学生语、数、外三科成绩的分布情况
    
要求：
    
(1) 每科使用不同颜色表示；

(2) 设置x轴、y轴有标题；
    
(3) 设置图表标题
    
    
数据源:成绩.xlsx

```python
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd

mpl.rcParams['font.sans-serif'] = ['FangSong']
mpl.rcParams['axes.unicode_minus'] = False

data = pd.read_excel('E:\项目文件\Python\可视化文档\成绩.xls').values

y1 = [150 if data[i][1] > 140 else data[i][1] // 10 * 10 for i in range(len(data))]
y2 = [250 if data[i][2] > 140 else data[i][2] // 10 * 10 + 100 for i in range(len(data))]
y3 = [350 if data[i][3] > 140 else data[i][3] // 10 * 10 + 200 for i in range(len(data))]

plt.figure(figsize=(20, 10))

d = 10
num_bin1 = int((max(y1) - min(y1)) // d)
num_bin2 = int((max(y2) - min(y2)) // d)
num_bin3 = int((max(y3) - min(y3)) // d)

plt.hist(y1, num_bin1, label='语文成绩')
plt.hist(y2, num_bin2, label='数学成绩')
plt.hist(y3, num_bin3, label='英语成绩')

x = [i for i in range(min(y1), max(y3) + 10, 10)]
plt.xticks(x, rotation=45)

plt.xlabel('成绩')
plt.ylabel('人数')
plt.title('学生三科成绩分析图')
#处理坐标轴
x = [i for i in range(min(y1), max(y1) + 10, 10)]
x += [i for i in range(min(y2), max(y2) + 10, 10)]
x += [i for i in range(min(y3), max(y3) + 10, 10)]
xticks = [i for i in range(min(y1), max(y1) + 10, 10)]
xticks += [i - 100 for i in range(min(y2), max(y2) + 10, 10)]
xticks += [i - 200 for i in range(min(y3), max(y3) + 10, 10)]
plt.xticks(x, xticks)

plt.legend(loc='best')

plt.grid(alpha=0.4)

plt.show()
```

## 绘制散点图

【例2-14】利用散点图分析不同种类(species)鸢尾花萼片(sepal_size)和花瓣(petal)大小相关性
    
要求：
    
(1) x轴表示sepal_size,y轴表示petal_size
    
(2) 不同种类鸢尾花使用不同颜色表示
    
数据源:iris-v.csv

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import pandas as pd

mpl.rcParams['font.sans-serif'] = ['FangSong']
mpl.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(20, 10), dpi=80)

data = pd.read_csv('E:\\项目文件\\Python\\可视化文档\\iris-v.csv').values


x1 = [data[i][-2] for i in range(len(data)) if data[i][-3] == 'setosa']
y1 = [data[i][-1] for i in range(len(data)) if data[i][-3] == 'setosa']
x2 = [data[i][-2] for i in range(len(data)) if data[i][-3] == 'versicolor']
y2 = [data[i][-1] for i in range(len(data)) if data[i][-3] == 'versicolor']
x3 = [data[i][-2] for i in range(len(data)) if data[i][-3] == 'virginica']
y3 = [data[i][-1] for i in range(len(data)) if data[i][-3] == 'virginica']

plt.figure(figsize=(55, 20))
plt.scatter(x1, y1, c='b', label='setosa')
plt.scatter(x2, y2, c='r', label='versicolor')
plt.scatter(x3, y3, c='g', label='virginica')
plt.xlabel('sepal_size')
plt.ylabel('petal_size')
minx = min(min(x1), min(x2), min(x3))
maxx = max(max(x1), max(x2), max(x3))
plt.xticks(np.arange(minx, maxx, 0.5), np.arange(minx, maxx, 0.5))
miny = min(min(y1), min(y2), min(y3))
maxy = max(max(y1), max(y2), max(y3))
plt.yticks(np.arange(miny, maxy, 0.5), np.arange(miny, maxy, 0.5))
plt.legend(loc='best')
plt.title('散点图分析不同种类(species)鸢尾花萼片(sepal_size)和花瓣(petal)大小相关性')
plt.show()
```

```python
from pyecharts import options as opts
from pyecharts.charts import Scatter
import pandas as pd

df = pd.read_csv('E:\\项目文件\\Python\\可视化文档\\iris-v.csv', header=None)
data = df.values

data = pd.read_csv('E:\\项目文件\\Python\\可视化文档\\iris-v.csv').values

x1 = [data[i][-2] for i in range(len(data)) if data[i][-3] == 'setosa']
y1 = [data[i][-1] for i in range(len(data)) if data[i][-3] == 'setosa']
x2 = [data[i][-2] for i in range(len(data)) if data[i][-3] == 'versicolor']
y2 = [data[i][-1] for i in range(len(data)) if data[i][-3] == 'versicolor']
x3 = [data[i][-2] for i in range(len(data)) if data[i][-3] == 'virginica']
y3 = [data[i][-1] for i in range(len(data)) if data[i][-3] == 'virginica']

scatter = Scatter()
scatter.add_xaxis(x1)
scatter.add_yaxis('setosa', y1) 
scatter.add_xaxis(x2)
scatter.add_yaxis('versicolor', y2)
scatter.add_xaxis(x3)
scatter.add_yaxis('virginica', y3)
scatter.set_global_opts(xaxis_opts=opts.AxisOpts(name='sepal_size', type_='value', name_gap=30, min_=9, max_=32),
                        yaxis_opts=opts.AxisOpts(name='petal_size', ),
                        title_opts=opts.TitleOpts(title='散点图分析不同种类(species)鸢尾花萼片(sepal_size)和花瓣(petal)大小相关性'),
                        legend_opts=opts.LegendOpts(
                            orient='vertical',
                            pos_top='15%',
                            pos_right='2%'
                        ),
                        )
scatter.set_series_opts(label_opts=opts.LabelOpts(is_show=False)) #不显示数据
# scatter.render(r"E:\项目文件\Python\render.html")
scatter.render_notebook()
```
